```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('data/Social_Network_Ads.csv')
data.head()
```

```python
features = data.iloc[:,1:-1].values
goals = data.iloc[:,-1].values
```

```python
from sklearn.preprocessing import LabelEncoder
features[:,0] = LabelEncoder().fit_transform(features[:,0])
```

```python
from sklearn.preprocessing import StandardScaler
features = StandardScaler().fit_transform(features)
pd.DataFrame(features)
```

```python
from sklearn.model_selection import train_test_split
input,test_input ,output,test_output = train_test_split(features,goals,test_size = .2, random_state = 0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5)
classifier.fit(input,output)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(test_output,classifier.predict(test_input))
pd.DataFrame(cm)
```

```python
for k in [1, 3, 5, 10, 20, 100, 150, len(input)]:
    classifier = KNeighborsClassifier(n_neighbors=k)
    classifier.fit(input,output)
    cm
    print('K = %d\n-------------'%k)
    print(pd.DataFrame(cm))
    print("model accuracy : {:.4f}".format(classifier.score(test_input,test_output)))
    print('-------------\n\n')
```
