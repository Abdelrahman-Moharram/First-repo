```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import LinearSVC
from sklearn.ensemble import VotingClassifier
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
import matplotlib.gridspec as gridspec
import itertools
%matplotlib inline

iris = datasets.load_iris()
X, y = iris.data[:, 1:3], iris.target

clf1 = LogisticRegression(random_state=1)
clf2 = LinearSVC(random_state=1)
clf3 = GaussianNB()

eclf = VotingClassifier(estimators=[('lr', clf1), ('svc', clf2), ('gnb', clf3)], voting='hard')

gs = gridspec.GridSpec(2, 2)

fig = plt.figure(figsize=(10,8))

for clf, lab, grd in zip([clf1, clf2, clf3, eclf], ['Logistic Regression', 'SVC', 'naive Bayes', 'Ensemble'],
                         itertools.product([0, 1], repeat=2)):
    clf.fit(X, y)
    ax = plt.subplot(gs[grd[0], grd[1]])
    fig = plot_decision_regions(X=X, y=y, clf=clf)
    plt.title("model {} with acu {:.2f}".format(lab, clf.score(X, y)))
```

```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
import matplotlib.gridspec as gridspec
import itertools
%matplotlib inline

iris = datasets.load_iris()
X, y = iris.data[:, 1:3], iris.target

clf1 = DecisionTreeClassifier(max_depth = 4)
clf2 = KNeighborsClassifier(n_neighbors = 7)
clf3 = SVC(kernel = 'rbf')
eclf = VotingClassifier(estimators=[('dt',clf1),("knn",clf2),('svc',clf3)])

gs = gridspec.GridSpec(2,2)

figure = plt.figure(figsize = (10,8))


for clf, lab, grd in zip([clf1, clf2, clf3, eclf], ['tree', 'KNN', 'SVC', 'Ensemble'], itertools.product([0, 1], repeat=2)):
    clf.fit(X, y)
    ax = plt.subplot(gs[grd[0], grd[1]])
    fig = plot_decision_regions(X=X, y=y, clf=clf)
    plt.title("model {} with acu {:.2f}".format(lab, clf.score(X, y)))
```
