```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
dataset = pd.read_csv('data/Mall_Customers.csv')
dataset.head()
```

```python
features = dataset.iloc[:, 1:].values

# solve categorical data
# solve the categorical data
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
features[:,0] = encoder.fit_transform(features[:,0])

pd.DataFrame(features).head()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3,random_state = 0)
kpredict = kmeans.fit_predict(features)
pd.DataFrame(kpredict)
```

```python
pd.DataFrame(features)
```

```python
kmeans.cluster_centers_[:,2]
```

```python
plt.scatter(features[kpredict == 0,-2],[features[kpredict == 0,-1]] ,s = 100,c = 'red', label = 'cluster1')
plt.scatter(features[kpredict == 1,-2],[features[kpredict == 1,-1]] ,s = 100,c = 'yellow', label = 'cluster2')
plt.scatter(features[kpredict == 2,-2], features[kpredict == 2, -1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:, -1], s = 300, c = '#333333', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
