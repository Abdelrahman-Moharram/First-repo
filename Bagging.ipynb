```python
from sklearn import datasets
from sklearn.ensemble import BaggingClassifier
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
import matplotlib.gridspec as gridspec
%matplotlib inline


iris = datasets.load_iris()
X = iris.data[:,1:3]
y = iris.target

bagging = BaggingClassifier(LogisticRegression(random_state = 0))
logistic = LogisticRegression(random_state = 0)

gs = gridspec.GridSpec(1,2)

figure = plt.figure(figsize = (10,8))

for clf , lab, grd in zip ([bagging,logistic],['logistic','Bagging'],[[0,0],[0,1]]):
    clf.fit(X,y)
    ax = plt.subplot(gs[grd[0],grd[1]])
    figure = plot_decision_regions(X, y, clf)
    plt.title("model {} with acu {:.2f}".format(lab, clf.score(X, y)))
```

```python
from sklearn.tree import DecisionTreeClassifier
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier
import matplotlib.gridspec as gridspec
%matplotlib inline

iris = datasets.load_iris()
X = iris.data[:,1:3]
y = iris.target

forest = RandomForestClassifier(n_estimators = 500,n_jobs = -1,random_state = 1)
tree = DecisionTreeClassifier(random_state = 1)

gs = gridspec.GridSpec(1,2)



fig = plt.figure(figsize = (10,8))


for clf, lab, grd in zip([forest,tree],['forest','tree'],[[0,0],[0,1]]):
    clf.fit(X,y)
    ax = plt.subplot(gs[grd[0], grd[1]])
    plot_decision_regions(X,y,clf)
    plt.title("The Model{} score is{:.2f}".format(lab, clf.score(X,y)))
```
