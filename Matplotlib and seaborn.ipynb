```python
# Fixing random state for reproducibility
np.random.seed(19680801)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
plt.hist(x, 50, density=True, facecolor='#333333', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.grid(True)
plt.show()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize=(15,10) )
axes = axes.flatten()
x = np.arange(-5,5,.1)
y = np.sin(x)
z = np.cos(x)
i = np.exp(x)
n = np.tan(x)
axes[0].plot(x,y,color = "red",label = "sin",linestyle = "-.")
axes[0].legend()
axes[0].set_title("sin")
axes[0].grid = True


axes[1].plot(x,z,color = "green",label = "cos",linestyle = "-")
axes[1].legend()
axes[1].set_title("cos")

axes[2].plot(x,n,color = "#333333",label = "tan",linestyle = "--")
axes[2].legend()
axes[2].set_title("tan")

axes[3].plot(x,i,color = "purple",label = "exp",linestyle = ":")
axes[3].legend()
axes[3].set_title("exp")
```
